```python
import pandas as pd
import numpy as np
import pygraphsort as pgs
import matplotlib.pyplot as plt
import seaborn as sns
```

## We are going to eliminate highly correlated features from our football dataset.

### We create the correlation matrix of our dataset and visualize it using a heatmap.

```python
data = pd.read_csv("Final.csv")

columns = data.columns[9:]

data = data[columns]

corr_matrix = data.corr().abs() 

plt.figure(figsize=(100, 100))

sns.heatmap(corr_matrix, xticklabels=False, yticklabels=False, cbar=False, square=True, annot=False, cmap='coolwarm')

plt.savefig("Heatmap_with_absvalue.png")

plt.show()




```

### We now use the Matrix Sorting Algorithm from Section $1.5$.

```python
G = corr_matrix.to_numpy()
G_sorted, P_sorter, P_ent = pgs.sort_matrix(G, epsilon=0.1, n_iter=20)

plt.figure(figsize=(100, 100))

sns.heatmap(G_sorted, xticklabels=False, yticklabels=False, cbar=False, square=True, annot=False, cmap='coolwarm')

plt.savefig("Heatmap_with_absvalue_sorted.png")

plt.show()
```

### Finally, we delete features by using an arbitrary parameter that came from looking at the heatmap.

```python
# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool_))

to_drop = [column for column in upper.columns if any(upper[column] > 0.568)]

data_reduced = data.drop(to_drop, axis=1)

print(f"Original number of features: {len(data.columns)}")
print(f"Reduced number of features: {data_reduced.shape[1]}")

corr_matrix = data_reduced.corr().abs()

G = corr_matrix.to_numpy()
G_sorted, P_sorter, P_ent = pgs.sort_matrix(G, epsilon=0.1, n_iter=20)

plt.figure(figsize=(100, 100))

sns.heatmap(G_sorted, xticklabels=False, yticklabels=False, cbar=False, square=True, annot=False, cmap='coolwarm')

plt.show()







```
